```python
#include<stdio.h>
int main()
{
	int r,n,s=0,temp;
	printf("enter the number");
	scanf("%d",&n);
	temp=n;
	while(n!=0)
	{
		r = n % 10;
		s = s * 10 + r;
	    n = n / 10;
    
	}
	if(s==temp)
	{
		printf("%d it is palindrome",s);
	}
	else
	{
		printf("%d it is not palindrome",s);
	}
	return 0;
}
```

```python
// READ THE TWO NUMBERS AS INPUT AND PRINT THE OUTPUT AS THE NUMBERS WHICH IS PERFECTLY DIVISIBLE BY 3 AND 5
// TEST CASES-- 1 15 -- 15, 15  45 -- 15 30 45
#include<stdio.h>
int main()
{
	int x,y;
	scanf("%d%d",&x,&y);
	while(x<=y)
	{
		if(x%3==0 && x%5==0)
		{
			printf("%d",x);
		x++;
	}
	return 0;
} 
```

```python
// READ THE NUMBER AS INPUT AND PRINT THE OUTPUT AS DIGITS OF THE NUMBER
// TEST CASES -- 145 - 5 4 1 ,1889 - 1 8 8 9
#include<stdio.h>
int main()
{
	int x,r;
	scanf("%d",&x);
	while(x!=0)
	{
		r = x % 10;
		printf("%d\t",r);
		x = x / 10;
	}
	return 0;
}
```
